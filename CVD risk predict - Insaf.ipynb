```python
import pandas as pd
import numpy as np

df = pd.read_csv("health_data.csv") #reads csv file and stores it in a pandas dataframe
df = df.drop_duplicates() #removes any duplicate data if present
df = df.drop(columns = "id") #drops the initial id column as unnamed is a better replacement for id
df.rename(columns={"Unnamed: 0": "id"}, inplace=True) #renames the column "Unnamed: 0" to id

df['age']= df['age']//365 #converts the age in days to age in years and rounds up.

m = {0:'M',1:'F'} #categorical gender values are mapped into M or F using a dictionary
df['gender'] = df['gender'].map(m) #The mapped gender values are replaced in columns using the map function

m1 = {0:'N',1:'Y'} #binary values are mapped into Y or N using a dictionary
df['smoke'] = df['smoke'].map(m1) #The mapped smoke values are replaced in columns using the map function
df['alco'] = df['alco'].map(m1) #The mapped alcohol intake values are replaced in columns using the map function
df['active'] = df['active'].map(m1) #The mapped physical activity values are replaced in columns using the map function
df['cardio'] = df['cardio'].map(m1)#The mapped presence of cvd disease values are replaced in columns using the map function


df
```
